```python
import requests
import numpy as np
import cv2
from PIL import Image
import os
from glob import glob
import json
from pdf2image import convert_from_path
import base64
from typing import Dict, List, Tuple
from datetime import datetime
import pandas as pd
def np_img_to_b64(img: np.array) -> str:
    
    _, buffer = cv2.imencode('.jpg', img)
    base64_byte_image = base64.b64encode(buffer)
    base64_str_image = base64_byte_image.decode("utf_8")
    
    return base64_str_image

def request_deepdoc(b64_str_img: str, url: str) -> Tuple[List, List, List]:
    image_payload = {'image': b64_str_img}
    response = requests.post(url, json = image_payload)
    response_json = json.loads(response.text)
    dfs_text = response_json["dataframes"]
    dfs = [pd.read_json(df_text) for df_text in dfs_text]
    return dfs
```

```python
table_segmentation_url = "https://ptvn-document-ai-deepdoctection-api-tzprdgk55a-as.a.run.app/api/table_segmentation"
path = "dataset/raw_po_cpall/5520fe2cf8ea4200a9658c1c275263ab.pdf"
images = convert_from_path(path)
```

```python
rgb_np_img = np.array(images[5])
bgr_np_img = cv2.cvtColor(rgb_np_img, cv2.COLOR_RGB2BGR)
base64_str_image = np_img_to_b64(bgr_np_img)
```

```python
dfs = request_deepdoc(base64_str_image, table_segmentation_url)
```
